```python
#Loading our Tables from the respective .Csv file
import pandas as pd

b = 'Telcom_dataset.csv'
pd1 = pd.read_csv(b, names= ['PRODUCT', 'VALUE', 'DATE_TIME', 'CELL_ON_SITE','DW_A_NUMBER_INT', 'DW_B_NUMBER_INT', 'COUNTRY_A', 'COUNTRY_B', 'CELL_ID', 'SITE_CODE'])
pd1 = pd1.drop([0])
pd1.head(5)
```

```python
#Loading our Tables from the respective .Csv file
c = 'Telcom_dataset2.csv'
pd2 = pd.read_csv(c ,names= ['PRODUCT', 'VALUE', 'DATE_TIME', 'CELL_ON_SITE','DW_A_NUMBER_INT', 'DW_B_NUMBER_INT', 'COUNTRY_A', 'COUNTRY_B', 'CELL_ID', 'SITE_CODE'])
pd2 = pd2.drop([0])
pd2.head()
```

```python
#Loading our Tables from the respective .Csv file
d = 'Telcom_dataset3.csv'
pd3 = pd.read_csv(d, names= ['PRODUCT', 'VALUE', 'DATE_TIME', 'CELL_ON_SITE','DW_A_NUMBER_INT', 'DW_B_NUMBER_INT', 'COUNTRY_A', 'COUNTRY_B', 'CELL_ID', 'SITE_CODE'])

pd3 = pd3.drop([0])
pd3.head()
```

```python
#Loading our Tables from the respective .Csv file
e = 'cells_geo.csv'
pd4 = pd.read_csv(e , delimiter= ';' )

pd4.head()
```

Merge the Telcom_datsets to one Dataset

```python
#Merging Telcom_dataset1, Telcom_dataset2,Telcom_dataset3
#separeted Date_time column to Time and Date column and the Dropped the Date_Time column 

pd_a = [pd1 ,pd2, pd3]
x =pd.concat(pd_a)
x.head()
x[['Date','Time']] = x.DATE_TIME.str.split(" ",expand=True)
x = x.drop(columns='DATE_TIME')
x.head()
```

##Merging the Telcom datasets with the Cells_geo dataset

```python

df_m = pd.merge(pd4, x, on='SITE_CODE', how= 'outer')
df_m.head(5)
df_m
```
